```c++

// create a vector to store price
std::vector<double> price;
price.clear();
// we know data has price on the 5th or 6th column, so just choose "col" = 5
int col = 5;
{
    std::ifstream input("test.csv");
    // use "i" to skip first row
    int i=0;
    while(!(input.eof()))
    {
        // skip row 0
        if(i==0){i++;continue;}
        
        // get row data separated by comma's
        auto rowData = MATH60082::ReadCSV::readRow(input);
        // now read in the fifth column as a "double"
        if(rowData.size()>col)
        {
            std::stringstream ss;
            ss << rowData[col];
            double temp;
            if( ss >> temp )
            {
                price.push_back(temp);
            }
            else
            {
                std::cout << "WARNING: could not read data" << std::endl;
            }
        }
    }
}
```

```c++
// print out data
for(int i=0;i<price.size();i++)
{
    std::cout << i << " " << price[i] << std::endl;
}
```
